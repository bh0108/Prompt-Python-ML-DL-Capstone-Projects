```python
# Call Main file to call 

#10.	Next, create the main project (.ipynb) file and import the car rental module in it.
#11.	Define the main method and create objects for both car rental and customer classes.
#12.	Inside the main method, take the customer’s input as a choice for displaying car availability, rental modes, or returning the cars.
#13.	Use the relevant method for the customer’s input and print relevant messages.
#14.	Run the main method to start your project.


from Car import CarRental
from CarRentalPlatform import Customer

def main():
    available_cars = {"Toyota": 10, "Honda": 8, "Tesla": 5}
    rental = CarRental(available_cars)
    customer = Customer("John")

    while True:
        print("\n1. Display available cars")
        print("2. Rent a car")
        print("3. Return a car")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            rental.display_available_cars()
        elif choice == "2":
            car = input("Enter the car you want to rent: ")
            num_cars = int(input("Enter the number of cars you want to rent: "))
            rental_mode = input("Enter rental mode (hourly/daily/weekly): ")
            rental_time = customer.request_cars(rental, car, num_cars, rental_mode)
        elif choice == "3":
            car = input("Enter the car you want to return: ")
            num_cars = int(input("Enter the number of cars you want to return: "))
            rental_mode = input("Enter rental mode (hourly/daily/weekly): ")
            customer.return_cars(rental, car, rental_time, rental_mode, num_cars)
        elif choice == "4":
            print("Thank you for using our car rental service.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
